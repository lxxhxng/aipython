## MISSION 1

```python
# • Numpy를 이용해서 생성한 값에 대한 산점도 그래프 그리기
# x축 값과 y축 값을 numpy의 arange( ), random.randint( ) 등을 사용해서
# 생성하고 그것을 여러 형식의 산점 도 그래프로 그려본다.
import numpy as np
x=np.arange(1,11)
y=np.random.randint(1,30,10)
plt.scatter(x, y, c='r', marker='^', label='random')
plt.legend()
plt.show()
```

## (3) 막대 그래프 : bar( )/barh( )

```python
x = ['A','B','C']
y = [100, 150, 120]

plt.bar(x,y)
plt.show()
```

## (4) 빈도 그래프(히스토그램) : hist( )

```python
x= [1,1,1,2,3,2,4]
plt.hist(x, bins=4, color='green', align='mid')
plt.show()
```

## (5) 그래프 제목, x축 제목, y축 제목 지정 : title( ), xlabel( ), ylabel( )

```python
x = ['A','B','C']
y = [100, 150, 120]
plt.bar(x,y)
plt.title('Test Graph')
plt.xlabel('Group')
plt.ylabel('Score')
plt.show()
```

## (6) 한 화면에 여러 개 그래프 나타내기 : subplots( )

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
fig, axs = plt.subplots(2,2)
axs[0,0].plot(x,y)
axs[0,1].scatter(x,y)
axs[1,0].bar(x,y)
axs[1,1].barh(x,y)
fig.suptitle('Drawing Graph')
plt.show()
```

## (1) 산점도 그래프 : scatterplot( )

```python
# sns.get_dataset_names()
# 이 함수를 호출하면 내장 데이터셋의 이름들이 출력되며
# 이를 통해 어떤 종류의 데이터셋들이 제공되고 있는지 확인할 수 있습니다.
# 이러한 내장 데이터셋들은 Seaborn의 기능과 특징을 살펴보고 연습하는 데 유용합니다.
```

```python
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=df_iris)
plt.title('Scatter Plot by seaborn', fontsize=20)
plt.show()
```

## (2) 범주별 데이터 개수를 표시하는 막대 그래프 : countplot( )

```python
sns.countplot(x='species', data=df_iris)
plt.show()
```

## (3) 범주별 항목값 평균을 표시하는 막대 그래프 : barplot( )

```python
sns.barplot(x='species',y='petal_length',data=df_iris, errorbar='sd')
plt.show()
```

## (4) 숫자값을 갖는 칼럼들의 상관관계를 보여주는 그래프 : pairplot( )

```python
sns.pairplot(df_iris, hue='species')
plt.show()
```

## (5) 산점도와 추세선(회귀선)을 동시에 보여주는 그래프 : regplot( )

```python
sns.regplot(df_iris, x='petal_length', y='petal_width')
plt.show()
```

## (6) 집계한 값에 따라 색깔을 표시하는 그래프 : heatmap( )

```python
df=df_iris.corr()
sns.heatmap(data=df,annot=True,fmt= '.2f',cbar=False)
plt.show()
```

## MISSION 3

```python
# 시본에서 제공하는 데이터인 ‘penguins’를 이용해서 그래프 그리기

df_penguins = pd.read_csv('/content/gdrive/MyDrive/겨울계절_파이썬기초/data_10일차/penguins.csv')
df_penguins
```

```python
# 1. load_dataset( )을 이용해서 penguins 데이터를 가져온 후
#    펭귄 품종(species)별 몸무게(body_mass_g)의 평균을 그 래프로 그리는데,
#    아래 그림과 같이 성별을 구분해서 그려보자.

sns.barplot(x='species', y='body_mass_g',data=df_penguins, errorbar='sd', hue='sex')
```

```python
# 2. penguins 데이터의 숫자값을 갖는 항목들끼리의 상관관계 그래프를 그려보자.

sns.pairplot(df_penguins, hue='species')
plt.show()

df_corr=df_penguins.corr()
sns.heatmap(data=df_corr,annot=True,fmt= '.2f',cbar=False)
plt.show()
```

```python
# 3. 2번 미션의 결과에서 칼럼들간의 상관관계를 확인하고 관계가 있는 두 칼럼을 선택하여 산점도와 추세선이 함께 나타나는
# 그래프를 그려보자.

sns.regplot(x='flipper_length_mm', y='body_mass_g', data=df_penguins)
plt.show()
```
