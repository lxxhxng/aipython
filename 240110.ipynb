# 데이터프레임의 결측치 처리

```python
df.isna().sum()
```

```python
# (2) 데이터프레임에서 결측치에 값 채우기

# 확인여부 칼럼의 결측치를 '완료'로 채우기 / 원본 df는 변경이 안됨
df.fillna({'확인여부':'완료'})
```
