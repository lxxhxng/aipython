```python
# 범주 단위 차이에 의한 가시화 불가능 상태
plt.plot(df.index, df['출생아 수'])
plt.plot(df.index, df['합계 출산율'])
plt.show()

# X축 공유
fig, ax1 = plt.subplots(figsize=(10, 7))
ax1.plot(df.index, df['출생아 수'], color='#ff812d') # 각각 다른색상 설정
ax2 = ax1.twinx() # X축을 공유하는 쌍둥이 axis 생성, Y범주 변경
ax2.plot(df.index, df['합계 출산율'], color='#ffd100') # 각각 다른색상 설정
```

```python
# 혼합형 그래프
fig, ax1 = plt.subplots(figsize=(10, 7))
ax1.set_ylabel('출생아 수 (천명)')
ax1.bar(df.index, df['출생아 수'], color='#ff812d')
ax2 = ax1.twinx()
ax2.set_ylabel('합계 출산율 (가임여성 1명당/명)')
ax2.plot(df.index, df['합계 출산율'], color='#ffd100')
plt.show()
```

```python
# 그래프 가시화 세부 작업
fig, ax1 = plt.subplots(figsize=(13, 5))
ax1.set_ylabel('출생아 수 (천명)')
ax1.set_ylim(200, 600)
ax1.set_yticks([200, 300, 400, 500])
ax1.bar(df.index, df['출생아 수'], color='#ff812d')
ax2 = ax1.twinx()
ax2.set_ylabel('합계 출산율 (가임여성 1명당/명)')
ax2.set_ylim(0, 1.5)
ax2.set_yticks([0, 1])
ax2.plot(df.index, df['합계 출산율'], color='#ffd100')
plt.show()
```

```python
# 그래프 가시화 세부 작업 : 중앙정렬 필요
fig, ax1 = plt.subplots(figsize=(13, 5))
ax1.set_ylabel('출생아 수 (천명)')
ax1.set_ylim(200, 600)
ax1.set_yticks([200, 300, 400, 500])
ax1.bar(df.index, df['출생아 수'], color='#ff812d')
for idx, val in enumerate(df['출생아 수']): # X좌표는 인덱스 0~ 넣고
    ax1.text(idx, val + 3, val, ha='center') # value값 막대 위, 중간에 표시

ax2 = ax1.twinx()
ax2.set_ylabel('합계 출산율 (가임여성 1명당/명)')
ax2.set_ylim(0, 1.5)
ax2.set_yticks([0, 1])
ax2.plot(df.index, df['합계 출산율'], color='#ffd100')
plt.show()
```

```python
# marker, ms, lw
fig, ax1 = plt.subplots(figsize=(13, 5))
ax1.set_ylabel('출생아 수 (천명)')
ax1.set_ylim(200, 600)
ax1.set_yticks([200, 300, 400, 500])
ax1.bar(df.index, df['출생아 수'], color='#ff812d')
for idx, val in enumerate(df['출생아 수']): # X좌표는 인덱스 0~ 넣고
    ax1.text(idx, val + 3, val, ha='center') # value값 막대 위, 중간에 표시

ax2 = ax1.twinx()
ax2.set_ylabel('합계 출산율 (가임여성 1명당/명)')
ax2.set_ylim(0, 1.5)
ax2.set_yticks([0, 1])
ax2.plot(df.index, df['합계 출산율'], color='#ffd100',
         marker='o', ms=10, lw=3)
plt.show()
```

```python
# markeredgecolor : 테두리 흰색 추가, markeredgewidth
fig, ax1 = plt.subplots(figsize=(13, 5))
ax1.set_ylabel('출생아 수 (천명)')
ax1.set_ylim(200, 600)
ax1.set_yticks([200, 300, 400, 500, 600])
ax1.bar(df.index, df['출생아 수'], color='#ff812d')
for idx, val in enumerate(df['출생아 수']): # X좌표는 인덱스 0~ 넣고
    ax1.text(idx, val + 3, val, ha='center') # value값 막대 위, 중간에 표시

ax2 = ax1.twinx()
ax2.set_ylabel('합계 출산율 (가임여성 1명당/명)')
ax2.set_ylim(0, 1.5)
ax2.set_yticks([0, 1])
ax2.plot(df.index, df['합계 출산율'], color='#ffd100',
         marker='o', mec='w', mew=3, ms=10, lw=3)
plt.show()
```

```python
# 최종본
fig, ax1 = plt.subplots(figsize=(13, 5))
fig.suptitle('출생아 수 및 합계출산율') # subplot 타이틀 코딩은 subtitle
ax1.set_ylabel('출생아 수 (천명)') # y축의 이름 설정
ax1.set_ylim(200, 600) # y축의 범위 설정
ax1.set_yticks([200, 300, 400, 500]) # y축의 눈금 설정
ax1.bar(df.index, df['출생아 수'], color='#ff812d')
for idx, val in enumerate(df['출생아 수']): # X좌표는 idx 변수에 0부터 넣고
    ax1.text(idx, val + 10, val, ha='center') # value 값을 막대 약간 위, 중앙에 표시

ax2 = ax1.twinx() # ax1과 X축을 공유하는 그래프를 그리겠다
ax2.set_ylabel('합계 출산율 (가임여성 1명당/명)')
ax2.set_ylim(0, 1.5)
ax2.set_yticks([0, 1])
ax2.plot(df.index, df['합계 출산율'], color='#ffd100',
        marker='o', mec='w', mew=3, ms=10, lw=3)
# marker: 점모양
# mec(marker edge color): 점 테두리(여기서는 힌색으로 구분해주고싶어서 w)
# mew: 점 테두리 굵기
# ms: 점 크기
# lw: 선 굵기

for idx, val in enumerate(df['합계 출산율']):
    ax2.text(idx, val + 0.05, val, ha='center')
plt.show()
```
