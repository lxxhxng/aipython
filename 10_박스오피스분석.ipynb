- **REST (Representational State Transfer) API**

  - 웹 서비스를 구현하는 아키텍처 스타일 중 하나
  - Client/Server 간 통신에 관한 일관된 interface를 제공하여 서버는 자원을 관리하고 클라이언트는 서버에서 해당 자원에 관한 상태를 요청

  - HTTP 프로토콜을 기반으로, URI (Uniform Resource Identifier: 자원에 대한 위치 정보), HTTP 메소드 (GET, POST, PUT, DELETE) 등을 이용해 클라이언트에서 서버로 데이터를 전송하고, 서버에서 클라이언트로 데이터를 반환


- **RESTful API**

  - REST API 구현 방법론 중 하나
  - RESTful API 구성요소:

    - **자원 (Resource):** URI로 표현되며, 클라이언트가 요청하는 정보
    - **행위 (Verb):** HTTP 프로토콜의 메소드를 사용하여 자원에 대한 행위
    - **표현 (Representation):** C/S 간 데이터를 주고받을 때 사용하는 표현 방식 (JSON, XML 등)


- **Open API**

  - 외부 개발자들이 특정 서비스나 플랫폼에서 제공하는 기능을 활용할 수 있도록 개방된 API
  - 일반적으로 RESTful API 형태로 제공되며, JSON이나 XML과 같은 데이터 형식으로 정보를 전송


- **request 모듈**

  - HTTP 요청을 보내는 데 사용하는 모듈

````python
# 함수 만들기
def boxoffice(tdt) :

    # OPEN API 데이터 가져오기1 (Url주소에 넣어서 요청)
    apikey = 'f5eef3421c602c6cb7ea224104795888'
    tdt = '20230101'
    url = f'http://kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?key={apikey}&targetDt={tdt}'
    resp = requests.get(url)

    # OPEN API 데이터 가져오기2 (요청값을 딕셔너리로 생성하여 요청)
    apikey = 'f5eef3421c602c6cb7ea224104795888'
    url = 'http://kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json'
    pram = {'key': apikey, 'targetDt': tdt} #tdt 타겟dt는 매개변수로 받음
    resp = requests.get(url, params=pram)

    data = resp.json() #제이슨으로 변환
    df = pd.DataFrame(data['boxOfficeResult']['dailyBoxOfficeList']) #변환된 데이터 추출 후, 데이터프레임 생성
    df = df[['openDt', 'rank', 'movieNm', 'salesAmt', 'salesAcc', 'audiCnt', 'audiAcc']] #필요한 열 추출
    df['showRange'] = data['boxOfficeResult']['showRange'] #조회일 추출
    df['showRange'] = df['showRange'].str[:4] + '-' + df['showRange'].str[4:6] + '-' + df['showRange'].str[6:8]

    df.columns = ['개봉일', '순위', '영화명', '매출액', '누적매출액', '관객수', '누적관객수', '조회일'] #한국어로 변환
    df = df.astype({'순위':'category', '매출액': int, '누적매출액': int, '관객수': int, '누적관객수': int})
    df['개봉일'] = pd.to_datetime(df['개봉일']) #열 데이터타입 변경
    df['조회일'] = pd.to_datetime(df['조회일'])
    df['요일'] = df['조회일'].dt.weekday
    df['요일명'] = df['요일'].map({0:'월', 1:'화', 2:'수', 3:'목', 4:'금', 5:'토', 6:'일'}) #요일 데이터 생성
    return df
````

````python
df = boxoffice('20230101') #특정요일 함수 호출
df
````

````python
# 2주간 데이터프레임 생성 : 최종
df = pd.DataFrame()
for i in range(1, 15) :
    tdt = f'202301{i:02d}'
    print(tdt)
    df = pd.concat([df, boxoffice(tdt)])
df.info()

df
````

````python
df.reset_index(inplace=True)

df
````

````python
# 관객수 상위 10개 영화 추출
df.sort_values(['관객수'], ascending=False).head(10)
````
